```python
import pandas as pd
import os
from icecream import ic
import warnings
warnings.filterwarnings('ignore')
```

```python
datapath="F:/insofe/Data/rawDataLoan/" #display all filenames in datapath
os.listdir(datapath)
```

```python
text_readme=open(datapath+'Readme.txt','r')#reading text file
f_readme=text_readme.readlines()
f_readme
```

```python
col_names=['CustomerID','WasTheLoanApproved']
df_results=pd.read_csv(datapath +'result.tsv',sep='\t',header=None,names=col_names)
df_results.head()
```

```python
col_app='CustomerID LoanPayoffPeriodInMonths LoanReason RequestedAmount InterestRate Co-Applicant'
df_app=pd.read_csv(datapath+'app2.tsv',sep='\t',header=None,names=col_app.split(' '))
df_app.head()
```

```python
col_borrower='CustomerID YearsAtCurrentEmployer YearsInCurrentResidence Age RentOrOwnHome TypeOfCurrentEmployment NumberOfDependantsIncludingSelf'
df_borrower=pd.read_csv(datapath+'borrower.tsv',sep='\t',header=None,names=col_borrower.split(' '))
df_borrower.head()
```

```python
col_credit='CustomerID CheckingAccountBalance DebtsPaid SavingsAccountBalance CurrentOpenLoanApplications'
df_credit=pd.read_csv(datapath+'credit.tsv',sep='\t',header=None,names=col_credit.split(' '))
df_credit.head()
```

```python
df_app['CustomerID']=df_app['CustomerID'].astype('object')
```

```python
df_merge_results_app=pd.merge(df_results,df_app,right_on='CustomerID',left_on='CustomerID')
df_merge_results_app
```

```python
df_merge_app_borrower=pd.merge(df_merge_results_app,df_borrower,on='CustomerID')
df_merge_app_borrower.head()
```
